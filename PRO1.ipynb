```python
# --- 4. Outlier Detection (Numeric Columns) using IQR Method ---
print("\n=== Outlier Count (IQR method) ===")

for col in numeric_cols:
    Q1 = fin[col].quantile(0.25)
    Q3 = fin[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = fin[(fin[col] < lower) | (fin[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
# --- 5. Check Inconsistent Formatting (extra spaces, mixed case) ---
print("\n=== Inconsistent Formatting Check ===")

for col in categorical_cols:
    trimmed = fin[col].str.strip()
    if not trimmed.equals(fin[col]):
        print(f"Inconsistent spacing found in column: {col}")

    if fin[col].str.lower().nunique() != fin[col].nunique():
        print(f"Case formatting inconsistency found in column: {col}")
```

# Handling Missing Values

```python
 # Missing Values Summary 
print(fin.isnull().sum())

# ---Separate numeric and categorical columns
num_cols = fin.select_dtypes(include=['int64', 'float64']).columns
cat_cols = fin.select_dtypes(include=['object']).columns

# Fill numeric missing values with median
fin[num_cols] = fin[num_cols].fillna(fin[num_cols].median())

#Fill categorical missing values with mode
for col in cat_cols:
    fin[col] = fin[col].fillna(fin[col].mode()[0])

# --- Check again ---
print("\nMissing values after handling:")
print(fin.isnull().sum())
```

**obesrvations:** All missing values in numeric columns were successfully filled using the median, ensuring the data remains stable and unaffected by outliers.
Categorical columns were imputed using the most frequent category (mode), preserving the original distribution of categories.
After imputation, the dataset contains zero missing values, making it ready for further analysis and modeling.

# Bi-Variate Analysis

```python
import matplotlib.pyplot as plt

# Group by Age and calculate mean Loan Amount
age_loan = fin.groupby('Age')['Loan Amount'].mean()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(age_loan)

plt.xlabel('Age')
plt.ylabel('Average Loan Amount')
plt.title('Average Loan Amount Across Different Ages')
plt.grid(True)

plt.show()
```

**Observation:** The average loan amount varies across ages, indicating that borrowers of different age groups have different borrowing patterns.

Certain age ranges show noticeably higher average loan amounts, suggesting those age groups may have greater financial needs or higher loan eligibility.
